## Working with a local catalog (..the name)

```python
fr = registry.FileRegistry(endpoint, cache=True)
```

```python
fr.get_catalog()
```

```python
pprint(fr.get_entries())
```

```python
fr_id1 = 'mms1_feeps_brst_electron'
fr_id2 = 'mms2_feeps_brst_electron'
start_date = '2020-02-01T00:00:00Z'
end_date =   '2020-02-02T00:00:00Z'
```

```python
#pprint(fr.get_catalog()['catalog'])
pprint(fr.get_entry(fr_id1))
```

```python
file_registry1 = fr.request_file_registry(fr_id1, start_date=start_date, end_date=end_date, overwrite=False)
```

```python
file_registry1
```

```python
print('Python Hash of File | Start Date | File Size')
fr.stream(file_registry1, lambda bo, d, f: print(hash(bo.read()), d.replace(' ', 'T')+'Z', f))
```

```python
file_registry2 = fr.request_file_registry(fr_id2, start_date=start_date, end_date=end_date, overwrite=False)
```
